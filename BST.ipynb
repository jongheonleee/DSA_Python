```python
class Node:
  def __init__(self, key):
    self.key = key
    self.parent = None
    self.left = None
    self.right = None

  def __str__(self):
    return str(self.key)

  def preorder(self):
    # M -> L -> R
    if self != None:
      print(self.key)

      if self.left:
        self.left.preorder()

      if self.right:
        self.right.preorder()

  def inorder(self):
    # L -> M -> R
    if self != None:
      if self.left:
        self.left.inorder()

      print(self.key)

      if self.right:
        self.right.inorder()

  
  def postorder(self):
    # L -> R -> M
    if self != None:
      if self.left:
        self.left.postorder()

      if self.right:
        self.right.postorder()

      print(self.key)

  # iterator 공부하고 다시 작성하기
  def __iter__(self):
    return self.preorder()

  def __next__(self):
    if self != None:
      raise StopIteration

    else:
      self.preorder()

    return str(self.key)


### BST
class BST:
  def __init__(self):
    self.root = None
    self.size = 0

  def __len__(self):
    return self.size

  def __iter__(self):
    return self.root.__iter__()

  def find_loc(self, key: int) -> Node:
    # if there is the key, retyrn the node with key
    # otherwise, return parent node

    if self.size == 0:
      return None

    p, v = None, self.root
    while v != None:
      if v.key == key:
        return v

      elif v.key < key:
        p, v = v, v.right

      else:
        p, v = v, v.left

    return p

  
  def search(self, key: int) -> Node:
    v = self.find_loc(key)

    if v == None:
      return None

    else:
      return v

  
  def insert(self, key: int) -> Node:
    p = self.find_loc(key)

    if p == None or p.key != key:
      v = Node(key)

      if p == None:
        self.root = v

      else:
        v.parent = p

        if p.key >= key:
          p.left = v

        else:
          p.right = v

      self.size += 1
      return v


    else:
      print("key is already in BST")
      return p


  def delete_by_merging(self, x: Node) -> None:
    a, b = x.left, x.right
    pt = x.parent
    c, m = None, None 

    if a != None:
      c, m = a, a

      while m.right:
        m = m.right

      if b != None:
        b.parent = m

      m.right = b

    else:
      c = b

    if pt != None:
      if c: 
        c.parent = pt

      if pt.key < c.key:
        pt.right = c

      else:
        pt.left = c

    else:
      self.root = c

      if c:
        c.parent = None

    self.size -= 1
    return None


  def delete_by_copying(self, x: Node) -> None:
    pass
    # i will write this later

  

```
