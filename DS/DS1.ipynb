```python
class Queue:
  def __init__(self, size:int) -> None:
    self.queue:list = [None for _ in range(size)]
    self.front:int = 0
    self.rear:int = 0
    self.size:int = size
    self.crnt_size:int = 0

  def __str__(self) -> str:
    ans = ''

    for i in range(self.front, self.rear):
      ans += f' [ {self.queue[i]} ]'

    return ans 

  def enqueue(self, val:int) -> None:
    if self.rear == self.size:
      print('queue is full')

    else:
      self.queue[self.rear] = val
      self.rear += 1

      self.crnt_size += 1

  def dequeue(self) -> int:
    if self.front == self.rear:
      print('queue is empty')

    else:
      x = self.queue[self.front]
      self.front += 1
      self.crnt_size -= 1

      return x

  def is_empty(self) -> bool:
    return self.front == self.rear and self.front == 0

  def is_full(self) -> bool:
    return self.rear == self.size

  
q = Queue(5)

q.enqueue(1)
q.enqueue(4)
q.enqueue(7)
q.enqueue(3)
q.enqueue(8)

print(q)

x1 = q.dequeue()
x2 = q.dequeue()

print(x1, x2, sep=' , ')

print(q)
```

```python
class Node:
  def __init__(self, val: int):
    self.val:int = val
    self.next:int = None


class LinkedList:
  def __init__(self) -> None:
    self.head = None
    self.size = 0

  def __str__(self) -> str:
    ans = ''

    crnt_node = self.head 
    while crnt_node:
      ans += f'{crnt_node.val} -> '
      crnt_node = crnt_node.next

    return ans

  def __len__(self) -> int:
    return self.size

  def push_front(self, val) -> None:
    new_node = Node(val)

    if self.head == None:
      self.head = new_node

    else:
      new_node.next = self.head
      self.head = new_node

    self.size += 1
    return None

  def push_back(self, val) -> None:
    new_node = Node(val)

    if self.head == None:
      self.head = new_node
      self.size += 1
      return None


    crnt_node = self.head
    while crnt_node.next:
      crnt_node = crnt_node.next

    crnt_node.next = new_node
    self.size += 1
    return None


  def pop_front(self) -> object:
    if self.head == None:
      print('LL is empty')
      return None

    else:
      x, next = self.head, self.head.next
      x.next = None
      self.head = next

      return x

  def pop_back(self) -> object:
    if self.head == None:
      print('LL is empty')
      return Node


    else:
      crnt_node = self.head

      while crnt_node.next:
        prev = crnt_node
        crnt_node = crnt_node.next

      prev.next = crnt_node.next
      return crnt_node

  
  def search(self, val) -> object:
    if self.head == None:
      return None

    crnt_node = self.head
    while crnt_node:
      if crnt_node.val == val:
        return crnt_node

      crnt_node = crnt_node.next

    return None

  def remove(self, val) -> None:
    if self.head == None:
      return None

    crnt_node = self.head
    while crnt_node.next:
      prev = crnt_node
      crnt_node = crnt_node.next

      if crnt_node.val == val:
        prev.next = crnt_node.next

    del crnt_node
    return None


        

    

ll = LinkedList()

for i in range(1, 6):
  ll.push_back(i)

print(ll)
print(ll.search(3).val)
print(ll.remove(3))
print(ll)
print(ll.pop_front().val)
print(ll)
print(ll.pop_back().val)
print(ll)
```

```python
class Node:
  def __init__(self, val):
    self.val = val
    self.next = None
    self.prev = None


class DoublyLinkedList:
  def __init__(self):
    self.head = Node(0)
    self.tail = Node(0)
    self.head.next, self.tail.prev = self.tail, self.head
    self.size = 0

  def __iter__(self):
    crnt_node = self.head

    while crnt_node:
      yield crnt_node
      crnt_node = crnt_node.next



  def __str__(self)-> object:
    return '<=>'.join(str(node.val) for node in self)

  def __len__(self) -> int:
    return self.size

  def push_front(self, val: int) -> None:
    new_node = Node(val)

    if self.size == 0:
      new_node.prev, new_node.next = self.head, self.tail
      self.head.next, self.tail.prev = new_node, new_node


    else:
      front_node = self.head.next
      new_node.prev, new_node.next = self.head, front_node
      self.head.next, front_node.prev = new_node, new_node


    self.size += 1
    
    return None

  def push_back(self, val: int) -> None:
    new_node = Node(val)

    if self.size == 0:
      new_node.prev, new_node.next = self.head, self.tail
      self.head.next, self.tail.prev = new_node, new_node

    else:
      back_node = self.tail.prev
      new_node.prev, new_node.next = back_node, self.tail
      back_node.next, self.tail.prev = new_node, new_node


    self.size += 1

  def pop_front(self) -> object: 
    if self.size == 0:
      return None

    else:
      target, next = self.head.next, self.head.next.next
      self.head.next, next.prev = next, self.head

      self.size -= 1
      return target

  def push(self, loc, val) -> None:
    if loc > self.size:
      return None

    new_node = Node(val)

    if loc <= self.size - loc:
      # it's better than starting at head
      crnt_node = self.head
      while crnt_node:
        loc -= 1
        crnt_node = crnt_node.next

        if loc == 0:
          break

    else:
      crnt_ndoe = self.tail

      while crnt_node:
        loc -= 1
        crnt_node = crnt_node.prev

        if loc == 0:
          break
    
    prev_node = crnt_node.prev
    new_node.prev, new_node.next = prev_node, crnt_node
    prev_node.next, crnt_node.prev = new_node, new_node

    self.size += 1

    return None
          

  
  def pop_back(self) -> object:
    if self.size == 0:
      return None

    else:
      target, prev = self.tail.prev, self.tail.prev.prev
      prev.next, self.tail.prev = self.tail, prev

      self.size -= 1
      return target

  def search(self, val) -> None:
    if self.size == 0:
      return None

    front_node = self.head 
    back_node = self.tail


    while front_node and back_node and front_node != back_node:
      if front_node.val == val:
        return front_node

      if back_node.val == val:
        return back_node

      front_node = front_node.next
      back_node = back_node.prev

    return None


  def remove(self, val) -> None:
    target = self.search(val)

    if target == None:
      return None

    prev, next = target.prev, target.next
    prev.next, next.prev = next, prev
  
    del target
    return None

  
dl = DoublyLinkedList()

for i in range(1, 10):
  dl.push_front(i)

print(dl)
print(dl.search(3).val)
dl.push_back(12)
print(dl)
dl.push(4, 15)
print(dl)
print(dl.size)

```
